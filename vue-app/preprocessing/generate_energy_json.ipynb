```python
import pandas as pd
import json
import numpy as np
```

```python
climate_data_cleaned = pd.read_csv("../data/climate_data_cleaned.csv")
climate_data_cleaned
```

```python
# filter for countries
countries = ['CAN', 'USA', 'AUT', 'DEU','ITA', 'GBR','JPN','CHN','IND','SAU']
climate_data_cleaned = climate_data_cleaned[climate_data_cleaned['country'].isin(countries)]
climate_data_cleaned
```

```python
# filter for relevant fields
climate_data_cleaned = climate_data_cleaned[['country', 'year', 'en_per_cap']]
climate_data_cleaned
```

```python
# round energy values to 2 decimal places
climate_data_cleaned['en_per_cap'] = climate_data_cleaned['en_per_cap'].round(decimals = 2)
# sort values by country name
climate_data_cleaned = climate_data_cleaned.sort_values(['country', 'year'])
climate_data_cleaned
```

```python
# convert to json
result = climate_data_cleaned.to_json(orient="records")
with open('../data/energy_per_cap.json', 'w') as f:
    f.write(result)
```
