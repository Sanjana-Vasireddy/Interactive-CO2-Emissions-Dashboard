```python
import pandas as pd
import json
import numpy as np
```

# Create JSON file from CSV

```python
geojson_mapped = pd.read_csv("../data/geojson_convertible.csv")
geojson_mapped
```

```python
# filter for countries
countries = ['CAN', 'USA', 'AUT', 'DEU','ITA', 'GBR','JPN','CHN','IND','SAU']
geojson_mapped = geojson_mapped[geojson_mapped['country'].isin(countries)]
geojson_mapped
```

```python
# convert year to string for labels
years = [str(i) for i in range(1991,2009)]
print(years)
```

```python
# convert data to long format
geojson_mapped_long = pd.melt(geojson_mapped, id_vars=['country', 'longitude', 'latitude'], value_vars=years,
       var_name='year', value_name='emissions')
geojson_mapped_long
```

```python
# round emissions values to 2 decimal places
geojson_mapped_long['emissions'] = geojson_mapped_long['emissions'].round(decimals = 2)
geojson_mapped_long
```

```python
# convert long format to json
result = geojson_mapped_long.to_json(orient="records")
with open('../data/emissions.json', 'w') as f:
    f.write(result)
```

# Convert JSON to GeoJSON

```python
data = json.load(open('../data/emissions.json'))
data
```

```python
geojson = {
    "type": "FeatureCollection",
    "features": [
    {
        "type": "Feature",
        "geometry" : {
            "type": "Point",
            "coordinates": [d["longitude"], d["latitude"]],
            },
        "properties" : d,
     } for d in data]
}

geojson
```

```python
output = open('../data/emissions.geojson', 'w')
json.dump(geojson, output)

print(geojson)
```
